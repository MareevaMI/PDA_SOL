```python
import pandas as pd
import requests,zipfile,io
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
med = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/insurance.csv')
```

```python
med.head(10)
```

```python
med.info
```

```python
sns.pairplot(med)
# Вывод: закономерность вывести по данному датасету трудно, но можно провести удачную кластеризацию
```

```python
sns.relplot(
    x="age",
    y="charges",
    data=med,
    kind='line',
    col='smoker')
   # Вывод 1: сумма расходов на медицинские услуги находится в прямой зависимости от возраста человека (чем старше человек, тем выше расходы)
   # Вывод 2: уровень расходов на медицинские услуги курильщиков выше, чем у некурящих людей
```

```python
plt.figure(figsize = (10,7))

plt.hist(med['smoker'], bins = 2)

plt.title('Число курящих')
plt.xlabel('')
plt.ylabel('Кол-во человек')

plt.show()

# Вывод: большее количество людей - не курит
```

```python
plt.figure(figsize = (10,7))

plt.scatter(med['age'], med['charges'])

plt.title('График разброса значений')
plt.xlabel('Возраст')
plt.ylabel('Сумма расходов')

plt.show()

# Вывод 1: Средний уровень расходов на медицинские услуги варьируется от 2 000 до 13 000
# Вывод 2: С увеличением возраста равномерно увеличивается уровень расходов
# Вывод 3: На основании полученного графика можно выделить более менее чёткие социальные классы (по финасовому признаку)
```

```python
data = med['children'].value_counts().reset_index()
data.head()
```

```python
plt.figure(figsize = (10,7))

plt.bar(data['index'],data['children'])

plt.title('Количество детей у опрошенных')
plt.xlabel('Кол-во детей')
plt.ylabel('Кол-во человек')
plt.xticks(rotation = 30)

plt.show()

# Вывод: большая часть людей из выборки не имеют детей
```

```python
data = med['smoker'].value_counts()
data.head()
```

```python
plt.figure(figsize = (10,7))

plt.pie(data, autopct='%1.1f%%')

plt.title('Доля курящих людей')
plt.legend(data.index)

plt.show()
```
